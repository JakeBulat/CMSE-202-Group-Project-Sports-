```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pro_stats_work = pd.read_csv("QBStats_all.csv")
pro_stats_work
```

```python
sorted_dataA = pro_stats_work.groupby(['qb','year']).sum()
sorted_dataA = sorted_dataA.drop(columns = ["ypa","rate"])

sorted_dataB = pro_stats_work.groupby(['qb','year']).mean()
sorted_dataB = sorted_dataB.drop(columns = ["att","cmp","yds","td","int","sack","loss","game_points"])

sorted_dataA["ypa"] = sorted_dataB["ypa"]
sorted_dataA["rate"] = sorted_dataB["rate"]

pro_stats = sorted_dataA.query("att > 350")
```
